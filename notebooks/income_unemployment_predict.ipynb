```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
income=pd.read_csv('../data/curated/income.csv')
```

```python
income
```

```python
income.columns
```

```python
sample=income.sample(n=10,random_state=114514)
```

```python
for i in range(len(sample)):
    plt.scatter([2015,2016,2017,2018,2019],sample.iloc[i,2:7])
    plt.show()
```

```python
# remove empty values
income = income[income['2014-15 #jobs*000'] != 'np']
income = income[income['2015-16 #jobs*000'] != 'np']
income = income[income['2016-17 #jobs*000'] != 'np']
income = income[income['2017-18 #jobs*000'] != 'np']
income = income[income['2018-19 #jobs*000'] != 'np']
income = income[income['2014-15 MEDIAN EMPLOYMENT INCOME PER JOB'] != 'np']
income = income[income['2015-16 MEDIAN EMPLOYMENT INCOME PER JOB'] != 'np']
income = income[income['2016-17 MEDIAN EMPLOYMENT INCOME PER JOB'] != 'np']
income = income[income['2017-18 MEDIAN EMPLOYMENT INCOME PER JOB'] != 'np']
income = income[income['2018-19 MEDIAN EMPLOYMENT INCOME PER JOB'] != 'np']
income = income.reset_index(drop=True)
income
```

```python
# by taking 10 samples from the number of jobs each year, we can see that , 
# they all follows a really linear relationship
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
lst2022=[]
lst2023=[]
lst2024=[]
lst2025=[]

X = np.array([2015,2016,2017,2018,2019]).reshape(-1, 1)  

for i in range(len(income)):
    
    y = list(income.iloc[i,2:7])  

    model = LinearRegression()
    model.fit(X, y)

    
    X_predict = np.array(2022).reshape(-1, 1)
    lst2022.append(float(model.predict(X_predict)))

    X_predict = np.array(2023).reshape(-1, 1)
    lst2023.append(float(model.predict(X_predict)))

    X_predict = np.array(2024).reshape(-1, 1)
    lst2024.append(float(model.predict(X_predict)))

    X_predict = np.array(2025).reshape(-1, 1)
    lst2025.append(float(model.predict(X_predict)))
```

```python
predict=income.iloc[:,:2]
predict["2022"]=lst2022
predict["2023"]=lst2023
predict["2024"]=lst2024
predict["2025"]=lst2025
predict
```

```python
predict.to_csv('../data/curated/income_num_predict.csv',index=False)
```

```python
dff=pd.read_csv('../data/curated/unemployment.csv')
dff
```

```python
dff.columns
```

```python
dff = dff[dff['Dec-10'] != '-']
dff = dff.reset_index(drop=True)
dff
```

```python
cols = ['Dec-10', 'Mar-11', 'Jun-11', 'Sep-11',
       'Dec-11', 'Mar-12', 'Jun-12', 'Sep-12', 'Dec-12', 'Mar-13', 'Jun-13',
       'Sep-13', 'Dec-13', 'Mar-14', 'Jun-14', 'Sep-14', 'Dec-14', 'Mar-15',
       'Jun-15', 'Sep-15', 'Dec-15', 'Mar-16', 'Jun-16', 'Sep-16', 'Dec-16',
       'Mar-17', 'Jun-17', 'Sep-17', 'Dec-17', 'Mar-18', 'Jun-18', 'Sep-18',
       'Dec-18', 'Mar-19', 'Jun-19', 'Sep-19', 'Dec-19', 'Mar-20', 'Jun-20',
       'Sep-20', 'Dec-20', 'Mar-21', 'Jun-21', 'Sep-21', 'Dec-21', 'Mar-22']
dff[cols] = dff[cols].apply(pd.to_numeric, errors='coerce', axis=1)
dff
```

```python
year=dff.iloc[:,2:4]
year.head()
```

```python
# 2010 and 2022 are not included since they only have one month of data
year['2011']=dff.iloc[:, 5:9].mean(axis=1)
year['2012']=dff.iloc[:, 9:13].mean(axis=1)
year['2013']=dff.iloc[:, 13:17].mean(axis=1)
year['2014']=dff.iloc[:, 17:21].mean(axis=1)
year['2015']=dff.iloc[:, 21:25].mean(axis=1)
year['2016']=dff.iloc[:, 25:29].mean(axis=1)
year['2017']=dff.iloc[:, 29:33].mean(axis=1)
year['2018']=dff.iloc[:, 33:37].mean(axis=1)
year['2019']=dff.iloc[:, 37:41].mean(axis=1)
year['2020']=dff.iloc[:, 41:45].mean(axis=1)
year['2021']=dff.iloc[:, 45:49].mean(axis=1)
```

```python
year
```

```python
lst2022=[]
lst2023=[]
lst2024=[]
lst2025=[]

X = np.array([2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]).reshape(-1, 1)  

for i in range(len(year)):
    
    y = list(year.iloc[i,2:13])  

    model = LinearRegression()
    model.fit(X, y)

    
    X_predict = np.array(2022).reshape(-1, 1)
    lst2022.append(float(model.predict(X_predict)))

    X_predict = np.array(2023).reshape(-1, 1)
    lst2023.append(float(model.predict(X_predict)))

    X_predict = np.array(2024).reshape(-1, 1)
    lst2024.append(float(model.predict(X_predict)))

    X_predict = np.array(2025).reshape(-1, 1)
    lst2025.append(float(model.predict(X_predict)))
```

```python
predict=year.iloc[:,:2]
predict["2022"]=lst2022
predict["2023"]=lst2023
predict["2024"]=lst2024
predict["2025"]=lst2025
predict
```

```python
predict.to_csv('../data/curated/unemployment_predict.csv',index=False)
```
