preprocess coordinates of train stations and shopping center

```python
import pandas as pd
import geopandas as gpd
```

```python
gdf = gpd.read_file("../data/raw/train.geojson")
```

```python
gdf
```

```python
coor=gdf['geometry'].centroid
```

```python
list_arrays = [ list((geom.xy[0][0], geom.xy[1][0])) for geom in coor ]

list_arrays[0]
```

```python
list_arrays
```

```python
df=pd.DataFrame(list_arrays)
```

```python
df.to_csv('../data/curated/train_coord.csv',index=False)
```

```python
df1=pd.read_csv("../data/curated/train_coord.csv")
df1
```

```python
gdf = gpd.read_file("../data/raw/shopping.geojson")
```

```python
coor=gdf['geometry'].centroid

list_arrays = [ list((geom.xy[0][0], geom.xy[1][0])) for geom in coor ]

list_arrays
```

```python
name=gdf['name']
name.head()
```

```python
df=pd.DataFrame(list_arrays)
df.head()
```

```python
df['name']=name

df.head()
```

```python
df['name'] = df['name'].replace([None],'null')
```

```python

df.to_csv('../data/curated/shop_coord.csv',index=False)

df1=pd.read_csv("../data/curated/shop_coord.csv")
df1
```
